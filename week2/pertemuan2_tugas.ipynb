#Deskripsi Tugas
Pada tugas pratikum ini Anda akan menggunakan data **"Wisconsin Breast Cancer"**. Data tersebut terdiri dari 569 data yang digunakan untuk mendiagnonis jenis kanker **Malignant (M)** dan **Benign (B)**. Tugas Anda adalah,

1. Pisahkan antara variabel yang dapat digunakan dan variabel yang tidak dapat digunakan.
2. Lakukan proses encoding pada kolom "diagnosis".
3. Lakukan proses standarisasi pada semua kolom yang memiliki nilai numerik.
4. Lakukan proses stratified split data untuk membuat data latih dan data uji dengan rasio 80:20.

```python
import pandas as pd

wbc = pd.read_csv('wbc.csv')
```

```python
wbc.info()
```

Karena data yang disediakan sudah tidak terdapat nilai null (na), dapat dilihat nilai non-null content semuanya sama, maka tidak perlu dilakukan data imputation

## 1. Memisahkan variable yang dapat digunakan

Pada data WBC ini, semua data dapat digunakan kecuali id dan Unnamed: 32

```python
# Hapus kolom yang tidak digunakan
wbc = wbc.drop(columns=['id', 'Unnamed: 32'])
```

```python
wbc.head()
```

# 2. Melakukan proses encoding pada kolom "diagnosis".

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder() # membuat objek dari LabelEncoder

wbc['diagnosis'] = le.fit_transform(wbc['diagnosis']) # proses encoding
```

Setelah dilakukan encoding pada kolom diagnosis menggunakan LabelEncoder, maka diagnosis bernilai B akan diganti nilainya menjadi 0, sedangkan diagnosis bernilai M akan diganti nilainya menjadi 1

```python
wbc.head()
```

 # 3. Proses standarisasi pada semua kolom yang memiliki nilai numerik.

```python
import numpy as np
from sklearn.preprocessing import StandardScaler

# Inisialisasi objek StandardScaler
scaler = StandardScaler()

# mencari semua kolom dengan tipe np.number
# referensi: https://stackoverflow.com/questions/51684585/how-to-get-numeric-column-names-in-pandas-dataframe
numeric_columns = wbc.select_dtypes([np.number]).columns
# kolom diagnosis tidak perlu distandarisasi karena bertipe ordinal
numeric_columns = numeric_columns.drop('diagnosis')

# # Lakukan standarisasi pada setiap kolom numerik
for col in numeric_columns:
    wbc[col] = scaler.fit_transform(wbc[[col]])
```

```python
wbc.head()
```

## 4. Lakukan proses stratified split data untuk membuat data latih dan data uji dengan rasio 80:20.

```python
# Split data
from sklearn.model_selection import train_test_split

wbc_train, wbc_test = train_test_split(wbc, test_size=0.2, random_state=0, stratify=wbc['diagnosis'])

print(f'Jumlah label data asli:\n{wbc.diagnosis.value_counts()}')
print(f'Jumlah label data train:\n{wbc_train.diagnosis.value_counts()}')
print(f'Jumlah label data test:\n{wbc_test.diagnosis.value_counts()}')
```
